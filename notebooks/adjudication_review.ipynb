# Study 1 — Adjudication Review

Human adjudication of dual-coder disagreements (this review **is** the
human adjudication step documented in Appendix B of the paper).

Workflow:
1. Run `python -m src.coding.run_coding` — it exports
   `data/study1_disagreements.csv` and stops if any are unresolved.
2. Work through each disagreement below: read both coders' labels and
   rationales, open the source URL if needed.
3. Record your decision in `data/study1_adjudications.csv`
   (columns: `candidate_id, final_label, rationale`). `final_label` is
   `L1`/`L2`/`L3`/`L4` or `excluded`. **Rationale is mandatory.**
4. Re-run `python -m src.coding.run_coding` to build the final dataset.

```python
from pathlib import Path
import pandas as pd

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
dis = pd.read_csv(ROOT / 'data' / 'study1_disagreements.csv')
print(f'{len(dis)} disagreements to adjudicate')
dis
```

```python
# Show one disagreement in full (change i to step through them)
i = 0
row = dis.iloc[i]
print(f"candidate : {row.candidate_id}")
print(f"system    : {row.system}  ({row.date})")
print(f"url       : {row.url}")
print(f"coder A   : {row.coder_a}  -- {row.rationale_a}")
print(f"coder B   : {row.coder_b}  -- {row.rationale_b}")
```

```python
# Helper: append a decision to data/study1_adjudications.csv
import csv

def adjudicate(candidate_id: str, final_label: str, rationale: str):
    assert final_label in ('L1', 'L2', 'L3', 'L4', 'excluded')
    assert rationale.strip(), 'rationale is mandatory'
    path = ROOT / 'data' / 'study1_adjudications.csv'
    exists = path.exists()
    with open(path, 'a', newline='', encoding='utf-8') as f:
        w = csv.writer(f, lineterminator='\n')
        if not exists:
            w.writerow(['candidate_id', 'final_label', 'rationale'])
        w.writerow([candidate_id, final_label, rationale])
    print(f'recorded: {candidate_id} -> {final_label}')

# Example:
# adjudicate('aiid-1028', 'L3', 'Oversight point existed (confirmation prompt) but lacked variety to absorb the action class.')
```
